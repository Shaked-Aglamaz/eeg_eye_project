### In the graph beneath we see gaze cordinates
(0,0) = top left, (1,1) = bottom right.

The eyes almost completly match. 
The screen is devided into 4, each one has a point in the middle. 
* Until time 70 the gaze is centered, as we see the cordinates are around (0.5, 0.5).
* In trials 1-3 the gaze is left, as we see the cordinates are around (0.0, 0.5).
* In trial 4 the gaze is centered again. 
* In the beginning of trial 5 the gaze is up, as we see the cordinates are around (0.5, 0.0).
