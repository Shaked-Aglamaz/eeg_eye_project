```python
import pyxdf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
n_trials = 150 
n_action = 2 
LOWER = .2
UPPER = .8 
rw = np.zeros(shape=(n_action, n_trials))
rw[:,0] = np.random.random(size=2).clip(LOWER,UPPER)
drift = .025
for t in range(1, n_trials):
    rw[:, t] = rw[:,t-1] + np.random.normal(0, drift, 2)
plt.title("Reward Probs")
plt.plot(rw[0])
plt.plot(rw[1], c='red')
plt.xlabel('trials')
plt.ylabel('prob')
plt.legend(["blue card", "red card"])  
```
